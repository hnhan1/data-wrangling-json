## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
from pandas.io.json import json_normalize
```

```python
json_df.mjtheme_namecode.head()
```

```python
# Load json file
json_1=json.load((open('data/world_bank_projects.json')))
```

```python
# Normalize and extract semi-structured data ('mjtheme_namecode') into a flat table
json_normlz=json_normalize(json_1,'mjtheme_namecode')
json_normlz.head(10)
```

```python
# Count 'code' 
counts=json_normlz['code'].value_counts()
counts.name='counts'
counts
```

```python
# Group by name and count 'name'
json_theme = json_normlz.groupby('name').size()
json_theme #122 records blank records
```

```python
# Sort 'name' count in descending order
json_sort = json_theme.sort_values(ascending=False).head(10)
json_sort 
```

```python
# Create dictionary and resety index
json_dict=json_normlz.sort_values('code',ascending=True).groupby('code').last().reset_index('code')
json_dict
```

```python
json_1=pd.DataFrame([json_dict['name'],counts])
json_1
```

```python
# Transpose dataframe and sort by 'counts'
json_1.transpose().sort_values('counts',ascending=False)
```

# 3. Create a dataframe with the missing names filled in.

```python
# Load json file
json_1=json.load((open('data/world_bank_projects.json')))

# Normalize and extract semi-structured data ('mjtheme_namecode') into a flat table
json_normlz=json_normalize(json_1,'mjtheme_namecode')
json_normlz.head(10)

```

```python
# Create dictionary for lookup
json_dict=json_normlz.sort_values('code',ascending=True).groupby('code').last().reset_index('code')
json_dict
```

```python
# Merge values back into original table for blanks with fieldname 'new_replacement'
c_values = pd.merge(json_normlz, json_dict, how='left', left_on='code', right_on='code', suffixes=['','_replacement'])

# Replace current values with new values
c_values.name = c_values.name_replacement

# Delete 'name_replacement' field
del c_values['name_replacement']

c_values
```
